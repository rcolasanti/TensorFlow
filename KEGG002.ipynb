```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from pprint import pprint

%matplotlib inline
```

```python
#full_data = pd.read_csv("datasets/wine/wine_train.csv")
full_data = pd.read_csv("datasets/KEGG/kegg_train.csv")
target = full_data[["I_Node"]]
atributes = full_data.drop("I_Node", axis=1) # drop labels for training set
atributes = atributes.drop("Pathway", axis=1) # drop labels for training set
atributes = atributes.drop("Isolated_Nodes", axis=1) # drop labels for training set
full_data.head()
```

```python
data_list = [
 'Characteristic_Path_Length',
 'Avg.num.Neighbours',
 'Density',
 'Heterogeneity',
 'Multi-edge_Node_Pair',
 'NeighborhoodConnectivity',
 'NumberOfDirectedEdges_real',
 'Stress',
 'Partner_Of_MultiEdged_NodePairs',
 'TopologicalCoefficient',
 'BetweennessCentrality',
 'Radiality',
 'Eccentricity',
 'ClosenessCentrality',
 'AverageShortestPathLength']
```

```python
atributes =  atributes[data_list]
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder(categories='auto')
target_hot = cat_encoder.fit_transform(target).toarray()
```

```python
#target = np.array(full_data["Cultivator"].copy(),dtype=int)
target = np.array(target["I_Node"],dtype=int)
pprint(target)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

attribs_labels = list(atributes)
pprint(attribs_labels)

# Replace any missing values with the median value
# Nomalize all the data

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('nomalize', MinMaxScaler()),
    ])


full_pipeline = ColumnTransformer([
        ("num", num_pipeline, attribs_labels),
#        ("cat", OneHotEncoder(), cat_attribs),
    ])

```

```python
attribs = full_pipeline.fit_transform(atributes)
```

```python
attribs
```

```python
dataset = {"attribs_labels":attribs_labels,"attribs":attribs.tolist(),"target":target.tolist(),"target_hot":target_hot.tolist()}
```

```python
#with open("datasets/wine/wine_train.json","w") as jout:
with open("datasets/KEGG/kegg_train.json","w") as jout:
    json.dump(dataset,jout)
```
