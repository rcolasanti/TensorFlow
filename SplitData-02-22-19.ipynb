```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint

%matplotlib inline

wine = pd.read_csv('datasets/wine/wine.csv', names = ["Cultivator", "Alchol", "Malic_Acid", "Ash", "Alcalinity_of_Ash", "Magnesium", "Total_phenols", "Falvanoids", "Nonflavanoid_phenols", "Proanthocyanins", "Color_intensity", "Hue", "OD280", "Proline"])

```

### Split data into a test and a training set
#### make sure that all classification types are represnted in the same ratios

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=7)
for train_index, test_index in split.split(wine, wine["Cultivator"]):
    strat_train_set = wine.loc[train_index]
    strat_test_set = wine.loc[test_index]
```

```python
strat_train_set.to_csv("datasets/wine/wine_train_7.csv",index=False)
strat_test_set.to_csv("datasets/wine/wine_test_7.csv",index=False)
```
